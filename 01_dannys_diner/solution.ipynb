```python
sales_menu_merge = pd.merge(sales, menu, on='product_id', how='left')
```

```python
sales_menu_merge
```

```python
all_merged = pd.merge(sales_menu_merge, members, on='customer_id', how='left')
```

```python
all_merged["order_date"] = pd.to_datetime(all_merged["order_date"],format="%d-%m-%Y")
all_merged["join_date"] = pd.to_datetime(all_merged["join_date"],format="%d-%m-%Y")
```

```python
all_merged
```

```python
solution1 = all_merged.groupby('customer_id')['price'].sum().reset_index()
```

```python
solution1
```

```python
solution2 = all_merged.groupby('customer_id')['order_date'].nunique().reset_index()
```

```python
solution2
```

```python
all_merged["first_date"] = (
    all_merged.groupby("customer_id")["order_date"]
              .transform("min"))
```

```python
solution3 = all_merged[
    all_merged["order_date"] == all_merged["first_date"]
]

print(solution3[["customer_id", "product_name"]])
```
